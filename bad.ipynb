```python
#計算所有大學(包括四技和五專)的男女比例

# 取得男生數量和女生數量
male = '男生計'  
selected_male = data[male]
column_sum_male = selected_male.sum()
female = '女生計'  
selected_female = data[female]
column_sum_female = selected_female.sum()

# 計算男女比例
total_students = column_sum_male + column_sum_female
male_percentage = (column_sum_male / total_students) * 100
female_percentage = (column_sum_female / total_students) * 100

# 將比例四捨五入到小數點第二位
male_percentage = round(male_percentage, 2)
female_percentage = round(female_percentage, 2)

print("男生數量：", column_sum_male)
print("女生數量：", column_sum_female)
print("男生比例：", male_percentage, "%")
print("女生比例：", female_percentage, "%")
```

```python
#計算台師大學生的人數佔所有學生的百分比

# 選擇學校名稱為"台師大"的行
taiwan_normal_university = data[data['學校名稱'] == '國立臺灣師範大學']

# 計算台師大的人數總計
total_students_count = taiwan_normal_university['總計'].sum()

# 使用特定的列名來選取該列的所有數字，假設你要搜尋的列名是 column_name
column_name = '總計'  # 請替換為你要搜尋的列名
selected_column = data[column_name]

# 計算該列的所有數字總和
column_sum = selected_column.sum()


# 計算台灣師範大學學生的比例
ntnu_percentage = (total_students_count / column_sum) * 100

# 將比例四捨五入到小數點第二位
ntnu_percentage = round(ntnu_percentage, 2)

print("台灣師範大學學生數量：", total_students_count)
print("所有大學學生數量：", column_sum)
print("台灣師範大學學生比例：", ntnu_percentage, "%")
```

```python
#國立大學佔所有大學的比例

# 刪除學校名稱中的重複項目，並計算所有以 "國立" 開頭的學校數量
unique_school_names = data['學校名稱'].unique()
national_count = sum(name.startswith('國立') for name in unique_school_names)

# 刪除學校名稱中的重複項目並計算不同學校的數量總計
不同學校數量 = data['學校名稱'].nunique()

# 計算比例
percentage = (national_count /不同學校數量  ) * 100

# 將比例四捨五入到小數點第二位
percentage = round(percentage, 2)

print("國立的學校數量：", national_count)
print("所有大學數量總計：", 不同學校數量)
print("國立大學的比例：", percentage, "%")





```
